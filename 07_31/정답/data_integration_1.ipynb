필요한 패키지를 import 합니다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
from pandas_datareader import data
```

병합의 대상이 되는 데이터 프레임을 생성합니다.

```python
data1 = pd.read_csv('./2014년 졸음운전 교통사고.csv',encoding='euc-kr') 
data2 = pd.read_csv('./2015년 졸음운전 교통사고.csv',encoding='euc-kr') 
data3 = pd.read_csv('./2016년 졸음운전 교통사고.csv',encoding='euc-kr')
```

자료에 있느 형태의 데이터 프레임이 될수 있도록 병합을 진행합니다. 
병합과 함께 '구분'에 해당되는 인덱스로 정렬이 될수 있도록 합니다. 
* DataFrame의 set_index 함수를 사용하여 원하는 컬럼 속성이 인덱스가 되도록 합니다
* DataFrame.set_index(keys, drop=True, append=False, inplace=False)

```python
data = pd.concat([data1,data2,data3])	# 데이터 병합
data.set_index('구분',inplace=True)		# 병합된 데이터의 인덱스 설정, 변경된 인덱스는 반영이 되어야 함 !
data
```

년도, 월별로 인덱스를 설정 (멀티인덱스)하고 데이터를 보이도록 설정

```python
ans2=data
ans2.index = pd.to_datetime(ans2.index, format='%Y년%m월')
ans2.set_index([ans2.index.year,ans2.index.month], inplace=True)
ans2.index.names = ['년도','월']
ans2
```
